```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
factorial(5)
```

```python
#fibonacci
def fib(n):
    if n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)


fib(8)
```

```python
def div(a,b):
    c=a/b
    return c
div(8,4)
#div(3,15)
```

```python
def smart_div(func):
    def inner(a,b):
        if a<b:
            a,b=b,a
        return func(a,b)
    return inner
```

```python
result=smart_div(div)
result(10,5)
```

```python
result(5,10)
```

TASK-4: To generate fibonacci number using generator function


```python
def fib_generator(n):
    a,b=0,1
    while True:
        yield a
        a,b=b,a+b
        
for i in fib_generator(6):
    print(i)
```

LAMBDA FUNCTION/ ANONYMOUS FUNCTION

```python
f=lambda i:i*i
f(5)
```

```python
##lambda function to add
add=lambda a,b:a+b
print(add(4,5))
```

```python
##largest between 2 numbers using lambda function 
largest=lambda a,b:a if a>b else b
largest(3,4)
```

```python
#filter, map, reduce

l=[0,2,3,4,5,6,8]
l1=list(filter(lambda x:x%2==0,l))
print(l1)
```
